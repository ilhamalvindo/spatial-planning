# RANDOM FOREST REGRESSOR MODEL

```python
df["Sub Zona"] = df["Sub Zona"].astype('category')
df["Sub Zona - Cat"] = df["Sub Zona"].cat.codes

df["Kecamatan"] = df["Kecamatan"].astype('category')
df["Kecamatan - Cat"] = df["Kecamatan"].cat.codes
```

```python
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


train_df, test_df = train_test_split(df, test_size=0.1, random_state=42)


features = ['Sub Zona - Cat', 'Kecamatan - Cat', 'Prakiraan Luas Persil (m2)', 'Tinggi bangunan (m)']
target1 = 'Jumlah Lantai'
target2 = 'KDB (%)'
target3 = 'KDH (%)'
target4 = 'KLB'

X_train = train_df[features]
y_train1 = train_df[target1]
y_train2 = train_df[target2]
y_train3 = train_df[target3]
y_train4 = train_df[target4]

model1 = RandomForestRegressor(n_estimators=50,
                               min_samples_split=2,
                               min_samples_leaf=2,
                               max_features='auto',
                               max_depth=30,
                               random_state=42)

model2 = RandomForestRegressor(n_estimators=350,
                               min_samples_split=2,
                               min_samples_leaf=1,
                               max_features='auto',
                               max_depth=20,
                               random_state=42)

model3 = RandomForestRegressor(n_estimators=450,
                               min_samples_split=2,
                               min_samples_leaf=2,
                               max_features='auto',
                               max_depth=30,
                               random_state=42)

model4 = RandomForestRegressor(n_estimators= 300,
                               min_samples_split=2,
                               min_samples_leaf=1,
                               max_features='auto',
                               max_depth=10,
                               random_state=42)

model1.fit(X_train, y_train1)
model2.fit(X_train, y_train2)
model3.fit(X_train, y_train3)
model4.fit(X_train, y_train4)

X_test = test_df[features]
y_test1 = test_df[target1]
y_test2 = test_df[target2]
y_test3 = test_df[target3]
y_test4 = test_df[target4]

y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred3 = model3.predict(X_test)
y_pred4 = model4.predict(X_test)

rmse1 = mean_squared_error(y_test1, y_pred1, squared=False)
rmse2 = mean_squared_error(y_test2, y_pred2, squared=False)
rmse3 = mean_squared_error(y_test3, y_pred3, squared=False)
rmse4 = mean_squared_error(y_test4, y_pred4, squared=False)

print('Root Mean Squared Error Jumlah Lantai:', rmse1)
print('Root Mean Squared Error KDB (%):', rmse2)
print('Root Mean Squared Error KDH (%):', rmse3)
print('Root Mean Squared Error KLB:', rmse4)



#---------------------------------------------------------------#
mse1 = mean_squared_error(y_test1, y_pred1)
mse2 = mean_squared_error(y_test2, y_pred2)
mse3 = mean_squared_error(y_test3, y_pred3)
mse4 = mean_squared_error(y_test4, y_pred4)

print('Mean Squared Error Jumlah Lantai:', mse1)
print('Mean Squared Error KDB (%):', mse2)
print('Mean Squared Error KDH (%):', mse3)
print('Mean Squared Error KLB:', mse4)

#---------------------------------------------------------------#
mae1 = mean_absolute_error(y_test1, y_pred1)
mae2 = mean_absolute_error(y_test2, y_pred2)
mae3 = mean_absolute_error(y_test3, y_pred3)
mae4 = mean_absolute_error(y_test4, y_pred4)

print('Mean Absolute Error Jumlah Lantai:', mae1)
print('Mean Absolute Error KDB (%):', mae2)
print('Mean Absolute Error KDH (%):', mse3)
print('Mean Absolute Error KLB:', mae4)

#---------------------------------------------------------------#
r21 = r2_score(y_test1, y_pred1)
r22 = r2_score(y_test2, y_pred2)
r23 = r2_score(y_test3, y_pred3)
r24 = r2_score(y_test4, y_pred4)

print('R-Squared Jumlah Lantai:', r21)
print('R-Squared Error KDB (%):', r22)
print('R-Squared KDH (%):', r23)
print('R-Squared KLB:', r24)
```
